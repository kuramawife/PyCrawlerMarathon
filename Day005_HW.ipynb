```python
data = json.loads(r.text)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('總共獲取',len(data),'筆資料')
print('每筆資料的欄位:', data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題:',d['title'],'貼文時間:',d['createdAt'],'留言人數:',d['commentCount'],'按讚人數:',d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
from statistics import mean
nondata = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
nondata=json.loads(nondata.text)
data1=[d['commentCount'] for d in data]
print('熱門文章 平均留言人數:', mean(data1))
data2=[d['likeCount'] for d in data]
print('熱門文章 平均按讚人數:', mean(data2))
print('非熱門文章 平均留言人數:', mean(n['commentCount'] for n in nondata))
print('非熱門文章 平均按讚人數:', mean(n['likeCount'] for n in nondata))
```
